```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img1 = cv2.imread('swan.pgm',0)



ret,l =cv2.threshold(img1,0,255,cv2.THRESH_OTSU)


th2 = cv2.adaptiveThreshold(img1,255,cv2.ADAPTIVE_THRESH_MEAN_C,
            cv2.THRESH_BINARY,11,2)
plt.imshow(img1,'gray')
plt.title("img1")
plt.xticks([]),plt.yticks([])
plt.show()



plt.imshow(l,'gray')
plt.title("otsu threshold")
plt.xticks([]),plt.yticks([])
plt.show()

plt.imshow(th2,'gray')
plt.title("adaptive threshold")
plt.xticks([]),plt.yticks([])
plt.show()





img2 = cv2.imread('flower.pgm',0)


plt.hist(img2.ravel(),256,[0,256]); plt.show()




ret,l2 =cv2.threshold(img2,0,255,cv2.THRESH_OTSU)

th3= cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_MEAN_C,
            cv2.THRESH_BINARY,11,2)


plt.imshow(l2,'gray')
plt.title("otsu threshold")
plt.xticks([]),plt.yticks([])
plt.show()

plt.imshow(th3,'gray')
plt.title("adaptive threshold")
plt.xticks([]),plt.yticks([])
plt.show()





img3 = cv2.imread('tools.pgm',0)

plt.hist(img3.ravel(),256,[0,256]); plt.show()




plt.imshow(img3,'gray')
plt.title("image equalize")
plt.xticks([]),plt.yticks([])
plt.show()


ret,l3 =cv2.threshold(img3,0,255,cv2.THRESH_OTSU)


th4 = cv2.adaptiveThreshold(img3,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)

plt.imshow(l3,'gray')
plt.title("otsu threshold")
plt.xticks([]),plt.yticks([])
plt.show()


plt.imshow(th4,'gray')
plt.title("adaptive threshold")
plt.xticks([]),plt.yticks([])
plt.show()
```
